```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#### This is the most basic plotting ever. we plot (x,y) so let's define them and then use matplotlib to plot it.

```python
x=[1,2,3]
y=[4,9,6]
#defined x and y

plt.plot(x,y) #plots the graph but usually in the background
plt.show() # brings it into the forefront.
```

 ## How to add title, and axis names?
 > This should be added after the plot is plotted and before plt.show()

```python
plt.plot(x,y)
plt.xlabel("x-axis variable")
plt.ylabel("y-axis variable")
plt.title("Name of the Graph")
plt.show()
```

 ## How to add subtitles; a smaller subheading?


```python

plt.title("Name of the Graph\ny v/s x") #after the main title added \n without space followed by the subheading again, without space after \n
plt.show()
```

 ## How to add legends: something which labels the graph?
> legends are used when there are more than one graph and we need to identify which is which.

> For this let's create one more graph.

```python
x2 =[1,2,3]
y2 =[20,15,23]
plt.plot(x,y, label ="first graph") # label is what we intend to call that graph
plt.plot(x2,y2, label ="second graph") 
plt.xlabel("x-axis variable")
plt.ylabel("y-axis variable")
plt.title("Graph Name\nGraphs 1 & 2")
plt.legend()
plt.show()
```

# <font color=red> BAR GRAPH <font/>
> Shows the frequency y at particular categories of x. 
> For example, say,a bakery sells 3 types of cakes: Chocolate, Black Forest and Truffle . To show the number of cakes sold in each category, say, in a month, we can use Bar graph. 

```python
x4 =['Chocolate','Black Forest','Truffle']
y4 =[25,29,15]
plt.bar(x4,y4) # plots bar chart
plt.xlabel('Types of cakes')
plt.ylabel('count ')
plt.title('Monthly sales count')
plt.show()
```

```python
x4 =['Chocolate','Black Forest','Truffle']
y4 =[25,29,15]
y5 =[14,16,8]
plt.bar(x4,y4, label='January', color='r') # plots bar chart with color red
plt.bar(x4,y5, label='February', color='c') # plots bar chart with color cyan
plt.xlabel('Types of cakes')
plt.ylabel('count')
plt.title('Monthly sales count')
plt.legend()
plt.show()
```

# <font color=red>HISTOGRAM <font/>
> Shows the distribution of a population
> For example, say, we have salary of various employees in an organization. we can group them into bins, starting from 20,000, upto 100,000. We use histogram fo that.

```python
salary=[23,32,78,56,34,29,78,89,54,66,73,48,96,82,38,49,57,69,73,85,99,43,79,80]
bins=[20,30,40,50,60,70,80,90,100]
plt.hist(salary,bins,histtype='bar',rwidth=0.95,color='c')
plt.xlabel('salary (in 1000s)')
plt.ylabel('count')
plt.title('Salary distribution')
plt.show()
```
