```python
df =df[["Adj. Open","Adj. High","Adj. Low","Adj. Close","Adj. Volume",]]
#Adj. Open and Adj. Close tells us whether the stock price went up or down.
#Adj. High and Adj. Low tells us about the volatility of the stock
#But a simple linear regression is not going to seek out that relation. It simply works through whatever we feed.
df
```

#NOTE.
> There is no use with the given column as it is, hence we are changing these into new features.
> * high_low_percentage = (Adj. High-Adj. Low)*100/Adj. Low. This gives us the percentage of volatility in the stock.
> * price_change_percentage = (Adj. Close-Adj. Open)*100/Adj.Open. This gives us the percentage of change in the stock price. positive value gives % increase and -ve value gives % decrease.

```python
df['high_low_percentage'] = (df['Adj. High'] - df['Adj. Low']) *100.0/ df['Adj. Low']
df['price_change_percentage'] = (df['Adj. Close'] - df['Adj. Open']) *100.0/ df['Adj. Open']
```
